# Variable aleatoria

Realicemos el experimento consistente en lanzar una moneda no cargada dos veces. Su  espacio muestral es 

$$\Omega =\{(c,c),(c,+),(+,c),(+,+)\}$$

donde todos los sucesos elementales son equiprobables.

Si consideramos $X$= "número de caras obtenidas en los dos lanzamientos", $X$ es una aplicación que asocia a cada resultado del espacio muestral un valor numérico

$$X:\Omega  \longrightarrow  \mathbb{R}$$ 
$$(+,+) \longrightarrow  0$$ 
$$(c,+) \longrightarrow 1$$
$$(+,c) \longrightarrow 1$$
$$(c,c) \longrightarrow 2$$

Además, cada uno de estos valores se toma con una cierta probabilidad inducida por la aleatoridad del fenómeno al que está asociado. Así, podemos escribir, por ejemplo:

$$P[X=0]=P[(+,+)]=\frac{1}{4}$$
$$P[X=1]=P[(c,+) \cup (+,c)]=\frac{1}{4}+\frac{1}{4}=\frac{1}{2}$$
$$P[X=2]=P[(c,c)]=\frac{1}{4}$$

La noción de **variable aleatoria** es la de **una función que asigna un valor numérico a cada suceso elemental**.

Sea $(\Omega ,\mathcal{S},P)$ un espacio probabilístico, se denomina **variable aleatoria** (v.a.) a una aplicación:

$$X: \Omega \longrightarrow \mathbb{R}$$
$$w \in \Omega \longrightarrow X(w)\in \mathbb{R}$$

# Variable aleatoria discreta

Una **variable aleatoria $X$ es discreta** si el conjunto de valores que puede tomar $X$ con probabilidad no nula es discreto.

Se conoce como **función de masa de probabilidad** o función de probabilidad de una variable aleatoria discreta $X$ que toma los valores $x_{1} \le x_{2} \le \ldots \le x_{n}$ con probabilidades no nulas a la función:

$$p_{_X}:\mathbb{R \rightarrow }\left[ 0,1\right]$$

definida  por:

$$p_{_X}(x)= \begin{cases} P[X=x_{k}], & \text{si } x=x_{k},\, k=1,2,\dots,n \\ 0, & \text{en otro caso.} \end{cases}$$

Sea $X$ una variable aleatoria discreta que toma los valores $x_{1} \le x_{2} \le \ldots \le x_{n}$ entonces se verifican las siguientes propiedades:

* $0\leq p_{_X}(x_{k})\leq 1$ para todo $k$.

* $\displaystyle\sum_k p_{_X}(x_{k})=1$.

Si consideramos $X$= "número de caras obtenidas en los dos lanzamientos", la **función masa de probabilidad** de la variable alestoria es


|$x_i$ | 0 | 1 | 2 |
|---|---|---|---|
|$p_{_X}(x_i)$ | 1/4 | 1/2 | 1/4 |

porque 

$$P[X=0]=P[(+,+)]=\frac{1}{4}$$
$$P[X=1]=P[(c,+) \cup (+,c)]=\frac{1}{4}+\frac{1}{4}=\frac{1}{2}$$
$$P[X=2]=P[(c,c)]=\frac{1}{4}$$

## Función de distribución

Se denomina función de distribución de una variable aleatoria $X$ a la función $F_{_X}$ definida como sigue:

$$F_{_X}: \mathbb{R} \longrightarrow \left[ 0,1\right]$$
$$F_{_X}(x)=P\left[ X\leq x\right],\text{ }\forall x \in \mathbb{R}.$$

La función de distribución de la variable aleatoria $X$ describe la acumulación de probabilidad por la variable a lo largo de la recta real. Tiene su antecedente en la frecuencia relativa acumulada.

En las variables aleatorias discretas se cumple que 

* $F_{_X}(x)=P[X \leq x]=\displaystyle\sum_{x_{k}\leq x}p_{_X}(x_{k})$.

* $p_{_X}(x_{k})=F_{_X}(x_{k})-F_{_X}(x_{k-1})$.


Para el lanzamiento de una moneda en el que medimos el número de caras obtenidas, calculemos primero $F_{_X}$ en los posibles valores de $X=\{0,1,2\}$:

$$F_{_X}(0)=P\left[ X\leq 0\right]=P\left[ X=0\right]={1}/{4}=0.25$$ 
$$F_{_X}(1)=P\left[X \leq 1 \right]=P\left[X=0 \right]+P\left[X=1\right]={3}/{4}=0.75$$
$$F_{_{X}}(2)=P\left[ X\leq 2\right]=P\left[ X=0\right] +P\left[ X=1\right] +P\left[ X=2\right] =1$$

Pero $F_{_X}$ está definida en todo el conjunto de los números reales, por tanto:

$$F_{_X}(x)=\begin{cases} 0, & \text{si } x < 0 \\ 0.25, & \text{si } 0 \leq x<1  \\ 0.75, & \text{si } 1 \leq x<2 \\ 1, & \text{si } x \geq 2 \end{cases}$$

```python
%pylab inline 
%config InlineBackend.figure_format = 'retina'
pylab.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')

from scipy import stats
from scipy.stats import rv_discrete

valores = np.array([0, 1, 2])
probabilidades = [0.25, 0.5, 0.25]

X = rv_discrete(values=(valores, probabilidades))
muestra = X.rvs(size=1000)
muestra_ordenada = np.sort(muestra)

x = muestra_ordenada
y = np.arange(1, len(x)+1) / len(x)

_ = plt.plot(x, y, marker='.')
_ = plt.xlabel('X')
_ = plt.ylabel('FdD empírica')
plt.margins(0.02)
plt.show()
```

# Características de la variable aleatoria. La esperanza matemática


Cierto establecimiento contabiliza el número de ventas diario de cierto producto. Los datos del último mes fueron:

|Número de ventas | 0 | 1 | 2 | 3 | 4 |
|-----------------| --|---|---|---|---|
|Número de días   | 10| 12| 5 |2  |1  |

Considerando la variable estadística $X=$ "Número de ventas diarios", puede considerarse la distribución de frecuencias:


|$x_i$ | 0 | 1 | 2 | 3 | 4|
|---|---|---|---|---|---|
|$n_i$ | 10 | 12 | 5 | 2 | 1 |
|$f_i$ | 1/3 | 2/5 | 1/6 | 1/15 | 1/30|

Para dicha distribución podemos calcular una serie de coeficientes como por ejemplo $\overline{x}$, $M_e$, $s^{2}$, etc. Estas medidas empíricas tienen su fundamento en las frecuencias observadas de los valores de la variable.

__Después de observar el comportamiento de dicha variable durante un número elevado de meses, las regularidades observadas en las frecuencias relativas permiten la definición de una distribución de probabilidad que trate de explicar el comportamiento futuro del fenómeno__.

|$x_i$ | 0 | 1 | 2 | 3 | 4 |
|---|---|---|---|---|---|
|$p_{_X}(x_i)$ | 1/3 | 2/5 | 1/6 | 1/15 | 1/30 |

De forma análoga podemos resumir los aspectos más relevantes de esta distribución mediante una serie
de medidas teóricas como por ejemplo la esperanza, la mediana, la varianza, etc. Así podemos relacionar conceptos como los que se muestran en la siguiente tabla:

    
|Medidas Empíricas | Medidas Teóricas |
|---|---|
|Frecuencia relativa | Probabilidad |
|Frecuencia relativa acumulada | Función de distribución | 
|Variable estadística | Variable aleatoria |
|Media aritmética $(\overline{x})$ | Esperanza matemática  $(\mu)$ |
|Varianza $(s^2)$ | Varianza $(\sigma^2)$ |


Para la __función masa de probabilidad__ 

|$x_i$ | 0 | 1 | 2 | 3 | 4 |
|---|---|---|---|---|---|
|$p_{_X}(x_i)$ | 1/3 | 2/5 | 1/6 | 1/15 | 1/30 |

podemos calcular la __esperanza matemática__, también llamada __media teórica__ o __media poblacional__, de la siguiente manera 

$$\mu = E[X] = 0 \cdot 1/3 + 1 \cdot 2/5 + 2 \cdot 1/6 + 3 \cdot 1/15 + 4 \cdot 1/30 = 1.0666$$

Del mismo modo, la __varianza teórica__ o __varianza poblacional__ se define como

\begin{align}
\sigma^2 = Var[X] = E[X^2] - (E[X])^2 = \\
0^2 \cdot 1/3 + 1^2 \cdot 2/5 + 2^2 \cdot 1/6 + 3^2 \cdot 1/15 + 4^2 \cdot 1/30 - (1.0666)^2 = 1.0622
\end{align}

La __desviación típica poblacional__ sería entonces

$$\sigma = \sqrt{\sigma^2} = \sqrt{1.062222} = 1.0306$$

```python
valores = np.array([0., 1., 2., 3., 4.])
probabilidades = [1./3, 2./5, 1./6, 1./15, 1./30]
X = rv_discrete(values=(valores, probabilidades))

print(X.mean())
print(X.var())
print(X.std())
```
